```python
#Injury Deaths

#Only 297 NA values. Will fill these with the average since there aren't that many NAs. 
#Also the standard deviation is around 24 so its not a huge deviation.

data_df['Injury.deaths'] 

#store the mean of the column to a variable name
injurydeath_mean = data_df['Injury.deaths'].mean()

#replace Nan Values with that median (here we use the inplace method)
data_df['Injury.deaths'].fillna(injurydeath_mean,inplace=True)
```

```python
# Unemployment & Uninsured 
#Both have only 6 values that are NA.  Filling these with average.

#store the mean of the column to a variable name
Unemployment_mean = data_df['Unemployment'].mean()

#replace Nan Values with that median (here we use the inplace method)
data_df['Unemployment'].fillna(Unemployment_mean,inplace=True)

#store the mean of the column to a variable name
Uninsured_mean = data_df['Uninsured'].mean()

#replace Nan Values with that median (here we use the inplace method)
data_df['Uninsured'].fillna(Uninsured_mean,inplace=True)
```

```python
#Diabetes & Adult.obesity                                                   5

# Diabetes is measuring the percent of diabetes found in the population. 
#Adult Obesity is measuring the percentage of the adult population who is obese'
# they both at 5 NA values, setting them both equal to the mean


#store the mean of the column to a variable name
Diabetes_mean = data_df['Diabetes'].mean()
Adultobesity_mean = data_df['Adult.obesity'].mean()

#replace Nan Values with that median (here we use the inplace method)
data_df['Diabetes'].fillna(Diabetes_mean,inplace=True)
data_df['Adult.obesity'].fillna(Adultobesity_mean,inplace=True)
```

```python
# Children.in.single.parent.households(7) , Child.Poverty.living.in.families.below.the.poverty.line (1)
# Preschool.Enrollment.Ratio.enrolled.ages.3.and.4 (8)


#store the mean of the column to a variable name
Childreninsingleparenthouseholds_mean = data_df['Children.in.single.parent.households'].mean()
ChildPovertylivinginfamiliesbelowthepovertyline_mean = data_df['Child.Poverty.living.in.families.below.the.poverty.line'].mean()
PreschoolEnrollmentRatioenrolledages3and4_mean = data_df['Preschool.Enrollment.Ratio.enrolled.ages.3.and.4'].mean()

#replace Nan Values with that median (here we use the inplace method)
data_df['Children.in.single.parent.households'].fillna(Childreninsingleparenthouseholds_mean,inplace=True)
data_df['Child.Poverty.living.in.families.below.the.poverty.line'].fillna(ChildPovertylivinginfamiliesbelowthepovertyline_mean,inplace=True)
data_df['Preschool.Enrollment.Ratio.enrolled.ages.3.and.4'].fillna(PreschoolEnrollmentRatioenrolledages3and4_mean,inplace=True)

```

### Question 3 - Augment the data

**Finally, augment the data so that each county contains the relative vote count for each party in each election. (e.g. For each county, if the Republicans got 2300 votes and the Democrats got 1900 votes, then the relative vote shares should be 54% and 46%.)**

```python
#Create a new DF to add relative values to votes df

df2 = pd.DataFrame(votes_df)


#Add new columns for Relative Democrat votes
df2['Democrats % 08'] = df2['Democrats 08 (Votes)']/(df2['Democrats 08 (Votes)']+ df2['Republicans 08 (Votes)'])
df2['Democrats % 12'] = df2['Democrats 12 (Votes)']/(df2['Democrats 12 (Votes)']+ df2['Republicans 12 (Votes)'])
df2['Democrats % 16'] = df2['Democrats 16 (Votes)']/(df2['Democrats 16 (Votes)']+ df2['Republicans 16 (Votes)'])


#Add new columns for Relative Republican votes
df2['Republicans % 08'] = df2['Republicans 08 (Votes)']/(df2['Democrats 08 (Votes)']+ df2['Republicans 08 (Votes)'])
df2['Republicans % 12'] = df2['Republicans 12 (Votes)']/(df2['Democrats 12 (Votes)']+ df2['Republicans 12 (Votes)'])
df2['Republicans % 16'] = df2['Republicans 16 (Votes)']/(df2['Democrats 16 (Votes)']+ df2['Republicans 16 (Votes)'])
                                                

df2.head()
```
